```python
def clean_and_save_songs(song_name, lines):
    name_without_ext = song_name.split(".")[0]
    new_name = name_without_ext + "_edited.txt"
    new_dir = os.getcwd() + "\\tmp\\"
    
    file = open(new_dir + new_name, 'w+', encoding='utf8')
    
    # count '\n' and insert {y} or {/y}
    newline_counter = 0
    num_lines = len(lines)
    for i in range(num_lines + 1):
        if i < num_lines:
            if lines[i].startswith("\n"):
                newline_counter += 1
                if newline_counter % 2 == 1:
                    file.write("\n{y}")
                else:
                    file.write("{/y}\n")
            else:
                file.write(lines[i])
        else:
            file.write("{/y}")
    file.close()
```

```python
for song_name in os.listdir("."):
#     print(song_name)
    if song_name.endswith(".txt"):
        lines = read_lines_from_text_file(filename=song_name)
        clean_and_save_songs(song_name, lines)
            
```
