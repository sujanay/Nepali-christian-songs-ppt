```python
import pandas as pd
import seaborn as sns
pd.set_option("display.max_column", 1000)
pd.set_option("display.max_row", 500)
```

```python
df_nkjv = pd.read_csv("bible_cleaned_formatted_nkjv_nlp.csv")
df_abbr = pd.read_csv("bible_abbreviations_final_nlp.csv")
```

```python
df_nkjv.head(3)
```

```python
df_abbr.head(3)
```

```python
df_merged = df_nkjv.merge(df_abbr, on="book_id", how="inner")
```

```python
df_merged.head(5)
```

```python
# aggregated to get verse per book
df_vpb = df_merged.groupby(["testament_id", "book_id", "book_name", "book_abbreviation"]).agg(
    {
        "chapter": "count"
    }
).reset_index().rename(columns={"chapter": "no. of verse"}).sort_values(
    by="no. of verse", ascending=False
).reset_index(drop=True)
df_vpb
```

```python
# aggregated to get chapter per book
df_cpb = df_merged.groupby(["book_id", "book_name"]).agg(
    {
        "chapter": "nunique"
    }
).reset_index().rename(columns={"chapter": "no. of chapter"}).sort_values(
    by="no. of chapter", ascending=False
).reset_index(drop=True)
df_cpb
```

```python
ax = sns.countplot(y="book_name", data=df_merged.sort_values(by="book_id", ascending=True))
```
