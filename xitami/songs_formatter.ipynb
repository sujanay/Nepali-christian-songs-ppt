```python
import os
import re
import shutil
```

```python
filepath = "\\nepali\\"
dest_filepath = "\\OpenLP format - JT\\temp\\"

nepali_songs_dir = os.getcwd() + filepath
dest_songs_dir = os.getcwd() + dest_filepath
print(nepali_songs_dir)
print(dest_songs_dir)
```

```python
os.listdir(nepali_songs_dir)
```

```python
for song_name in os.listdir(nepali_songs_dir):
    bhajan_chorus = re.findall("[BC]\d\d\d", song_name)  # match B005 or C060
    if bhajan_chorus:
        tmp = song_name.split(" - ")[0]
        new_name = bhajan_chorus[0] + " - " + tmp + ".txt"
        print(new_name)
        shutil.copy(nepali_songs_dir + song_name, dest_songs_dir + new_name)
```
