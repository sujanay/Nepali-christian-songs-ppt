```python
import os
import uuid

import numpy as np
import pandas as pd

bhajan_chorus_path = "../first_round_formatting/"
```

```python
content_list = []
for filename in os.listdir(bhajan_chorus_path):
    if filename.endswith(".txt"):
        # find the number and song name with the filename
        bhajan_chorus_num = int(filename.split(" - ")[0].strip()[1:])
        bhajan_chorus_name = filename.split(" - ")[1].strip()[:-4]

        # print(bhajan_chorus_num)
        # print(bhajan_chorus_name)
        # open the file and parse the text
        with open(bhajan_chorus_path + filename) as f:
            lines = f.readlines()

        temp_content = ""
        for line in lines:
            line = line.replace("\n", "<br>")
            temp_content += line
        
        # add tuple to list for creating pandas dataframe
        # COLUMNS TO MATCH in the excel 
        # https://docs.google.com/spreadsheets/d/1m7lChx_ZGy9QqE2QxrT_RoQ0oLNN92znMSbrtRRjNyk
        # id	type	typeLong	title	scale	timeSignature	content content_uuid
        content_id = uuid.uuid4()
        content_type = 1 if filename.upper().startswith("B") else 2 if filename.upper().startswith("C") else 3
        content_type_long = "Bhajan" if content_type == 1 else "Chorus" if content_type == 2 else "Baal Chorus"
        content_title = bhajan_chorus_name
        content_scale = ""
        content_time_signature = ""
        content = temp_content
        
        content_list.append(
            (
                bhajan_chorus_num,
                content_type,
                content_type_long,
                content_title,
                content_scale,
                content_time_signature,
                content,
                content_id
            )
        )
        

columns = ["id", "type", "typeLong", "title", "scale", "timeSignature", "content", "uuid"]
data = pd.DataFrame(content_list, columns=columns)
data
```

```python
data.to_csv("bhajan_chorus_first_round.csv", index=False)
```
